# 00. Opacus Fundamentals - Concepts and Privacy Engine

Welcome to the first notebook of the Opacus learning journey. 

Opacus is an open-source Python library developed by Meta AI that enables training PyTorch models with **Differential Privacy (DP)**. It does this by modifying the standard Stochastic Gradient Descent (SGD) algorithm into **DP-SGD** (Differentially Private SGD). 

### What you will learn in this notebook:
1. The core philosophy of DP-SGD in Opacus.
2. Essential library imports.
3. The working unit of Opacus: The `PrivacyEngine`.
4. How to wrap standard PyTorch objects to make them private.

```python
# Run this to install opacus if you haven't already in your environment
# !pip install opacus
```

## 1. Standard PyTorch Imports
Before we touch Opacus, we need standard PyTorch components: 
- a model, 
- an optimizer, and 
- a dataloader.

Let's import the essentials and create some dummy data to simulate a training environment.

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset

# 1. Create dummy data (100 samples, 10 features each)
X = torch.randn(100, 10)
y = torch.randint(0, 2, (100,))
dataset = TensorDataset(X, y)
dataloader = DataLoader(dataset, batch_size=10)

# 2. Create a very simple PyTorch model
model = nn.Sequential(
    nn.Linear(10, 16),
    nn.ReLU(),
    nn.Linear(16, 2)
)

# 3. Standard Optimizer
optimizer = optim.SGD(model.parameters(), lr=0.01)
```

## 2. Enter the **"PrivacyEngine"**

In standard PyTorch, the optimizer computes gradients and updates weights. In DP-SGD, two extra steps must happen during the backward pass:
1. **Per-Sample Gradient Clipping:** Ensure no single data point has too much influence on the gradient.
2. **Noise Addition:** Add random noise (usually Gaussian) to the gradients before updating the weights.

Opacus abstracts all of this math away using the `PrivacyEngine`.

```python
from opacus import PrivacyEngine

# Initialize the engine
privacy_engine = PrivacyEngine()
```

## 3. The '_make_private()_' Method

The core operation of Opacus is taking your standard PyTorch objects and wrapping them in private equivalents. We pass our model, optimizer, and dataloader into `privacy_engine.make_private()`.

We also define three crucial hyper-parameters here:
* **`noise_multiplier`**: Determines how much noise is added to the gradients.
* **`max_grad_norm`**: The maximum L2 norm of the per-sample gradients (clipping threshold).
* **`epochs`** (optional but recommended): Helps the engine track the privacy budget over time.

```python
model_dp, optimizer_dp, dataloader_dp = privacy_engine.make_private(
    module=model,
    optimizer=optimizer,
    data_loader=dataloader,
    noise_multiplier=1.0,
    max_grad_norm=1.0,
)

print("Model wrapped:", type(model_dp))
print("Optimizer wrapped:", type(optimizer_dp))
print("DataLoader wrapped:", type(dataloader_dp))
```

**Notice the types!** Your standard `DataLoader` is now a `DPDataLoader`, and your `SGD` optimizer is a `DPOptimizer`. These wrappers automatically handle clipping and noising under the hood when you call `loss.backward()` and `optimizer.step()`.

## 4. Tracking the Budget (Epsilon, $\epsilon$)
Differential Privacy is measured in `epsilon` ($\epsilon$) and `delta` ($\delta$).
- $\epsilon$ (Epsilon): The **privacy budget**. Lower means more private, but usually less accurate.
- $\delta$ (Delta): The **probability of the privacy guarantee failing**. Usually set to $\frac{1}{dataset\_size}$.

The `PrivacyEngine` keeps a running tally of your $\epsilon$ spend. Let's see how much budget we've spent (which is 0 right now, since we haven't trained!).

```python
# 1. Take a single batch from our private dataloader
data, target = next(iter(dataloader_dp))

# 2. Standard forward and backward pass
optimizer_dp.zero_grad()
output = model_dp(data)
loss = nn.CrossEntropyLoss()(output, target)
loss.backward()

# 3. Take a single step! 
#(This applies clipping, adds noise, and logs the step in the accountant)
optimizer_dp.step()
```

```python
# 4. Now we can safely check the budget
# Delta should be strictly less than 1/N (N = size of dataset)
# We have 100 samples, so delta = 1e-3 is safe.
target_delta = 1e-3

epsilon = privacy_engine.get_epsilon(delta=target_delta)
print(f"Current Epsilon spent after 1 step: {epsilon:.4f}")
```

We have successfully initialized the `PrivacyEngine`, wrapped standard PyTorch components, and learned how to query the privacy budget.
